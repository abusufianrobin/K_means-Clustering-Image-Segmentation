```python
from skimage import io
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image = io.imread('ladybug.jpg')
```

```python
# Show the original image
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()

# Reshape the image
rows, cols, channels = image.shape
image_reshaped = image.reshape(rows * cols, channels)
```

For K = 8

```python
# Apply K-means clustering with K = 8
kmeans_8 = KMeans(n_clusters=8, random_state=42)
kmeans_8.fit(image_reshaped)

# Get the labels and cluster centers
labels_8 = kmeans_8.labels_
centers_8 = kmeans_8.cluster_centers_.astype(np.uint8)
```

```python
print("Labels Are: ",labels_8)
```

```python
print("Centers Are: ",centers_8)
```

```python
# Create the compressed image
compressed_image_8 = centers_8[labels_8].reshape(rows, cols, channels)
```

```python
# Show the compressed image for K = 8
plt.figure(figsize=(6, 6))
plt.imshow(compressed_image_8)
plt.title('Compressed Image (K = 8)')
plt.axis('off')
plt.show()
```

For K = 16

```python
# Apply K-means clustering with K = 16
kmeans_16 = KMeans(n_clusters=16, random_state=42)
kmeans_16.fit(image_reshaped)

# Get the labels and cluster centers
labels_16 = kmeans_16.labels_
print("Labels Are: ",labels_16)

centers_16 = kmeans_16.cluster_centers_.astype(np.uint8)
print("Centers Are: ",centers_16)

# Create the compressed image
compressed_image_16 = centers_16[labels_16].reshape(rows, cols, channels)

```

```python
# Show the compressed image for K = 16
plt.figure(figsize=(6, 6))
plt.imshow(compressed_image_16)
plt.title('Compressed Image (K = 16)')
plt.axis('off')
plt.show()
```

For K = 32

```python
# Apply K-means clustering with K = 32
kmeans_32 = KMeans(n_clusters=32, random_state=42)
kmeans_32.fit(image_reshaped)

# Get the labels and cluster centers
labels_32 = kmeans_32.labels_
centers_32 = kmeans_32.cluster_centers_.astype(np.uint8)

# Create the compressed image
compressed_image_32 = centers_32[labels_32].reshape(rows, cols, channels)

print("Labels Are: ",labels_32)
print("Centers Are: ",centers_32)
```

```python
# Show the compressed image for K = 32
plt.figure(figsize=(6, 6))
plt.imshow(compressed_image_32)
plt.title('Compressed Image (K = 32)')
plt.axis('off')
plt.show()
```
